```python
import random
count=0
for i in range(4):
    num1=random.randint(1,10)
    print(num1)
    num2=eval(input("enter number between 1-10"))
    if num1==num2:
        print("won")
        break
    else:
        print("Fail")
        count+=1
        if count==4:
            print(f"you exceeded 4 chances please try again after 24 hours")
        else:
            print(f"you still have {4-count} chances")
```

```python
#wap ask the user get 5 random numbers
# get it is an even number or odd number 
# also count how many even numbers are there 
# and count how many odd numbers are there

#idea 
# take two counters one even and odd count
# for loop 5 times 
# each time take random numbers 
# if condition
#    True counter update 
# else
#    update the counter

import random
even_count,odd_count=0,0
for i in range (5):
    num=random.randint(1,20)
    if num%2==0:
        #print('even')
        even_count=even_count+1
        
    else:
        odd_count=odd_count+1
print(f"the total number of even are: {even_count}")
print(f"the total number of odd are: {odd_count}")  
```

```python
#wap ask the user get 5 random numbers
# get it is an even number or odd number 
# also count how many even numbers are there 
# and count how many odd numbers are there

#idea 
# take two counters one even and odd count
# for loop 5 times 
# each time take random numbers 
# if condition
#    True counter update 
# else
#    update the counter
import random
def count1():
    even_count,odd_count=0,0
    even_sum,odd_sum=0,0
       
    for i in range (5):
        num=random.randint(1,20)
        if num%2==0:
            even_count=even_count+1
            even_sum=even_sum+num
        else:
            odd_count=odd_count+1
            odd_sum=odd_sum+num
    return (even_count,odd_count,even_sum,odd_sum)
```

```python
# create a func for above code 
# return all 4 
# print all 4
even_count,odd_count,even_sum,odd_sum=count1()
print(f"the total number of even are: {even_count}")
print(f"the total number of odd are: {odd_count}") 
print(f"the summation of all even numbers are: {even_sum}")
print(f"the summation of all add numbers are: {odd_sum}")
```
